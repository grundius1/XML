```python
import csv
import xml.etree.ElementTree as ET
```

```python
file = "C:/Users/bllanos/Desktop/project1/prueba1/Abril_2022/Operaciones/Defensa nuevo/masivooperacionesDEF_BALMISXL.csv"

print(file)
```

```python
with open(file) as f:
    csv_reader = csv.DictReader(f,delimiter=";")
    root = ET.Element("operaciones")
    for row in csv_reader:
        operacion = ET.SubElement(root,"operacion")
        intevencioncreated = False
        financiacioncreated = False
        territoriocreated = False
        mecanismoscreated = False
        actividadcreated = False
        nutcreated = False
        geloccreated = False
        indicadorescreated = False
        indicadoresresultados = False
        #print(row)
        for k, v in row.items():
            if v == "":
                pass
            else:
                if "cod_campo_intervencion" in k:
                    if intevencioncreated == False:
                        intervenciones = ET.SubElement(operacion,"campos_intervencion")
                        intervencion = ET.SubElement(intervenciones,"campo_intervencion")
                        intevencioncreated = True
                    if "porcentaje" in k:
                        porcentaje = ET.SubElement(intervencion,"porcentaje")
                        porcentaje.text = f"{v}"
                    else:
                        cod_intervencion = ET.SubElement(intervencion,f"{k}")
                        cod_intervencion.text = f"{v}"
                elif "cod_tipo_financiacion" in k:
                    if financiacioncreated == False:
                        financiaciones = ET.SubElement(operacion,"tipos_financiacion")
                        financiacion = ET.SubElement(financiaciones,"tipo_financiacion")
                        financiacioncreated = True
                    if "porcentaje" in k:
                        porcentaje = ET.SubElement(financiacion,"porcentaje")
                        porcentaje.text = f"{v}"
                    else:
                        cod_financiacion = ET.SubElement(financiacion,f"{k}")
                        cod_financiacion.text = f"{v}"
                elif "cod_dimension_terr" in k:
                    if territoriocreated == False:
                        territorios = ET.SubElement(operacion,"tipos_territorio")
                        territorio = ET.SubElement(territorios,"tipo_territorio")
                        territoriocreated = True
                    if "porcentaje" in k:
                        porcentaje = ET.SubElement(territorio,"porcentaje")
                        porcentaje.text = f"{v}"
                    else:
                        cod_territorio = ET.SubElement(territorio,f"{k}")
                        cod_territorio.text = f"{v}"
                elif "cod_mecanismo_entrega" in k:
                    if mecanismoscreated == False:
                        mecanismos = ET.SubElement(operacion,"mecanismos_entrega")
                        mecanismo = ET.SubElement(mecanismos,"mecanismo_entrega")
                        mecanismoscreated = True
                    if "porcentaje" in k:
                        porcentaje = ET.SubElement(mecanismo,"porcentaje")
                        porcentaje.text = f"{v}"
                    else:
                        cod_mecanismo = ET.SubElement(mecanismo,f"{k}")
                        cod_mecanismo.text = f"{v}"
                elif "cod_actividad" in k:
                    if actividadcreated == False:
                        actividades = ET.SubElement(operacion,"actividades_economicas")
                        actividad = ET.SubElement(actividades,"actividad_economica")
                        actividadcreated = True
                    if "porcentaje" in k:
                        porcentaje = ET.SubElement(actividad,"porcentaje")
                        porcentaje.text = f"{v}"
                    else:
                        cod_mecanismo = ET.SubElement(actividad,f"{k}")
                        cod_mecanismo.text = f"{v}"
                elif "cod_nut" in k:
                    if nutcreated == False:
                        nuts = ET.SubElement(operacion,"localizaciones_operacion")
                        nut = ET.SubElement(nuts,"localizacion_operacion")
                        nutcreated = True
                    if "porcentaje" in k:
                        porcentaje = ET.SubElement(nut,"porcentaje")
                        porcentaje.text = f"{v}"
                    else:
                        cod_mecanismo = ET.SubElement(nut,f"{k}")
                        cod_mecanismo.text = f"{v}"
                elif "cod_contrato" in k:
                    contratos = ET.SubElement(operacion,"contratos")
                    contrato = ET.SubElement(contratos,"contrato")
                    cod_contrato = ET.SubElement(contrato,f"{k}")
                    cod_contrato.text = f"{v}"
                    '''
                elif "cod_municipio" in k:
                    if geloccreated == False:
                        gelocs = ET.SubElement(operacion,"municipios")
                        geoloc = ET.SubElement(gelocs,"municipio")
                        geloccreated = True
                    if "cod_postal" in k:
                        porcentaje = ET.SubElement(geoloc,"cod_postal")
                        porcentaje.text = f"{v}"
                    else:
                        cod_mecanismo = ET.SubElement(geoloc,f"{k}")
                        cod_mecanismo.text = f"{v}"
                '''
                elif "cod_ind_productividad" in k:
                    #print(k)
                    if indicadorescreated == False:
                        indicadores = ET.SubElement(operacion,"indicadores_productividad")
                        #indicador = ET.SubElement(indicadores,"indicador_resultado")
                        indicadorescreated = True
                    if "cod_ind_productividad" in k:
                        if "val" in k:
                            elemento = ET.SubElement(indicador,"val_2014")
                            elemento.text = "0"
                            elemento = ET.SubElement(indicador,"val_2015")
                            elemento.text = "0"
                            elemento = ET.SubElement(indicador,"val_2016")
                            elemento.text = "0"
                            elemento = ET.SubElement(indicador,"val_2017")
                            elemento.text = "0"
                            elemento = ET.SubElement(indicador,"val_2018")
                            elemento.text = "0"
                            elemento = ET.SubElement(indicador,"val_2019")
                            elemento.text = "0"
                            elemento = ET.SubElement(indicador,"val_2020")
                            elemento.text = "0"
                            elemento = ET.SubElement(indicador,"val_2021")
                            elemento.text = "0"
                            elemento = ET.SubElement(indicador,"val_2022")
                            elemento.text = "0"
                            val = ET.SubElement(indicador,f"{k.split('-')[1]}")
                            val.text = f"{v.replace(',','.')}"
                        else:
                            indicador = ET.SubElement(indicadores,"indicador_productividad")
                            cod_ind = ET.SubElement(indicador,"cod_ind_productividad")
                            cod_ind.text = f"{v.split('-')[0]}"
                elif "cod_ind_resultado" in k:
                    if indicadoresresultados == False:
                        indcadoresresultados = ET.SubElement(operacion,"indicadores_resultados")
                        indicadoresresultados = True
                    indicadoresresultado = ET.SubElement(indcadoresresultados,"indicador_resultado")
                    cod_ind_resultado = ET.SubElement(indicadoresresultado,f"{k}")
                    cod_ind_resultado.text = f"{v}"
                    
                else:
                    elemento = ET.SubElement(operacion,f"{k}")
                    elemento.text = f"{v.replace(',','.')}"
                
        #break         
tree = ET.ElementTree(root)
tree.write("C:/Users/bllanos/Desktop/project1/prueba1/Abril_2022/Operaciones/Defensa nuevo/masivooperacionesDEF_BALMISXL.xml", encoding = "UTF-8", xml_declaration=True,)
```
