```python
def csv_reader(file, delimiter=";",endpath ="C:/Users/bllanos/Desktop/project1/prueba1/"):
    with open(file) as csv_file:
        csv_reader = csv.DictReader(csv_file,delimiter=delimiter)
        line_count = 0
        root = ET.Element("operacionesModif")
        for row in csv_reader:
            #print(row)
            operacion = ET.SubElement(root, "operacion")
            for k,v in row.items():
                #print(k,v)
                ET.SubElement(operacion, f"{k}").text = f"{v}"
            line_count += 1
        print(f'Processed {line_count} lines.')
    tree = ET.ElementTree(root)
    tree.write(endpath+"test2.xml")
```

```python
%%time
csv_reader(file_path)
```
