```python
def orgs_dic(file):
    organismos = {}
    tree = ET.parse(file)
    root = tree.getroot()
    for organismo in root:
        for item in organismo:
            if item.tag == "codOrganismo":
                v = item.text
            if item.tag == "NIF":
                k = item.text
        organismos.update({k:v})
    return organismos
```
