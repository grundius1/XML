```python
def get_num_contratos(filepath):
    df = pd.read_excel(filepath, sheet_name= "1. DATOS BÁSICOS CONTRATACIÓN")
    df.dropna(thresh=5,inplace=True)
    df.columns = df.iloc[0]
    df.drop(df.index[0],inplace=True)
    df.reset_index(inplace=True,drop=True)
    #display(df)
    #print(df.dtypes)
    if df.shape[0] != 0:
        df = df[df.elegible != "NO"]
        df = df[df.elegible != "N"]
        df = df[df.elegible != "n"]
        df = df[df.elegible != "n0"]
    #df.drop("NaN", axis=1, inplace=True)
    return df.shape[0]
```

```python
def get_df_contratos(filepath):
    df = pd.read_excel(filepath, sheet_name= "1. DATOS BÁSICOS CONTRATACIÓN")
    df.dropna(thresh=5,inplace=True)
    df.columns = df.iloc[0]
    df.drop(df.index[0],inplace=True)
    df.reset_index(inplace=True,drop=True)
    #display(df)
    #print(df.dtypes)
    if df.shape[0] != 0:
        df = df[df.elegible != "NO"]
        df = df[df.elegible != "N"]
        df = df[df.elegible != "n"]
        df = df[df.elegible != "n0"]
    #df.drop("NaN", axis=1, inplace=True)
    return df
```

```python
def get_validation_code(validacion):
    validacion = unicodedata.normalize("NFKD", validacion).encode("ascii","ignore").decode("ascii").strip()
    if validacion.casefold().lower() == "si" or validacion.casefold().lower() == "s":
        return "S"
    elif validacion.casefold().lower() == "no" or validacion.casefold().lower() == "n":
        return "N"
    elif validacion.casefold().lower() == "no procede":
        return "NP"
    else:
        raise ValueError(validacion, "no es un codigo valido")
```

```python
def ayuntamientos_con_el_fichero(aytsdic):
    aytos_sin_fichero = {}
    for k,v in aytsdic.items():
        if "Justificacion_FSUE-DANA.xlsx" not in os.listdir(os.path.join(fichero_padre,k)):
            aytos_sin_fichero.update({k: os.listdir(os.path.join(fichero_padre,k))[4:]})
    return aytos_sin_fichero
```
