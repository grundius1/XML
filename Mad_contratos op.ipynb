```python
import xml.etree.ElementTree as ET
```

```python
file = "C:/Users/bllanos/Desktop/project1/MAD-contratos/operacionesXml.xml"
```

```python
tree = ET.parse(file)
root = tree.getroot()
```

```python
operaciones = {}
for operacion in list(root):
    for campo in list(operacion):
        #rint(campo.tag)
        if campo.tag == "cod_operacion":
            key = campo.text
            values = []
        elif campo.tag == "contratos":
            for contrato in list(campo):
                for contrato_datos in list(contrato):
                    values.append(contrato_datos.text)
    operaciones.update({key:values})
    
```

```python
def flatenner(lista):
    contratosstring=""
    for item in lista:
        contratosstring += str(item)+";"
    return contratosstring
```

```python
import csv
with open('C:/Users/bllanos/Desktop/project1/MAD-contratos/operacion-contrato.csv', 'w') as f:
    for key in operaciones.keys():
        f.write("%s;%s\n"%(key,flatenner(operaciones[key])))
```
