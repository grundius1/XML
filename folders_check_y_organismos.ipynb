```python
def get_df_facturas(filepath):
    df = pd.read_excel(filepath, sheet_name= "2. DATOS BÁSICOS FACTURAS")
    df.dropna(thresh=6,inplace=True)
    df.columns = df.iloc[0]
    df.drop(df.index[0],inplace=True)
    df = df[df.elegible != "NO"]
    df = df[df.elegible != "N"]
    df.reset_index(inplace=True,drop=True)
    return df
```

```python
def get_df_contratos(filepath):
    df = pd.read_excel(filepath, sheet_name= "1. DATOS BÁSICOS CONTRATACIÓN")
    df.dropna(thresh=5,inplace=True)
    df.columns = df.iloc[0]
    df.drop(df.index[0],inplace=True)
    df.reset_index(inplace=True,drop=True)
    return df
```

def operations_info(folder):
    '''
    funcion que usa el CSV de descarga de operaciones de Dana para poder crear una lista de diccionarios
    con los datos de las operaciones:
        operacion
        beneficiario
        nombre de la operacion
    args: el folder donde se encuentra la carpeta de datso con el fichero Operaciones_XXX.csv
    '''
    file = os.path.join(folder,"data_blp","Operaciones_PRE.csv")
    operationslist =[]
    with open(file,encoding="utf-8") as f:
        f.readline()
        for line in f.readlines():
            dic={"operacion":"",
                    "benef":"",
                    "nombre":""}
            #print(line.split(";"))

            dic["nombre"] = line.split(";")[3].replace("=\"","").replace("\"","").replace("-"," ")
