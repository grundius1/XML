```python
def organismosdic(folder):
        path = os.path.join(mainfolder,"data_blp/ORGANISMOS.csv")
        dic = {}
        with open(path,encoding="utf8") as f:
            f.readline()
            for line in f.readlines():
                fields = line.split(";")
                try:
                    dic[fields[1]].append((fields[0],fields[2]))
                except KeyError:
                    dic.update({fields[1]:[(fields[0],fields[2].strip())]})
        for k,v in dic.items():
            if len(v) > 1:
                dic.update({k:list(set(v))})
        return dic
```
