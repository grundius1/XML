```python
def dateconverter(df):
    for column in df.columns:
        if isinstance(column, str):
            if column.startswith("Fecha"):
                #print(df[f"{column}"])
                df[f"{column}"] = pd.to_datetime(df[f"{column}"]).dt.strftime("%d/%m/%Y")
    return df
```

```python
def get_df_facturas(filepath):
    df = pd.read_excel(filepath, sheet_name= "2. DATOS BÁSICOS FACTURAS")
    df.dropna(thresh=6,inplace=True)
    df.columns = df.iloc[0]
    df.drop(df.index[0],inplace=True)
    df = df[df.elegible != "NO"]
    df = df[df.elegible != "N"]
    df = df[df.elegible != "n"]
    df = df[df.elegible != "no"]
    df.reset_index(inplace=True,drop=True)
    df = dateconverter(df)
    return df
```

```python
def operation_getter(ayuntamiento):
    return ayuntamientos[ayuntamiento][0]
```

```python
def benefieciario_getter(ayuntamiento):
    return ayuntamientos[ayuntamiento][2]
```

```python
def get_subtipologia(concepto):
    if concepto.startswith("a)"):
        return "31a"
    elif concepto.startswith("b)"):
        return "31b"
    elif concepto.startswith("c)"):
        return "31c"
    elif concepto.startswith("d)"):
        return "31d"
    else:
        print(concepto)
        raise ValueError ( "el concepto de gasto esta mal redactado")
```

```python
def get_contract(ayto,contractcode):
    contrats = pd.read_csv(os.path.join(fichero_padre,"data_blp/correspcontratos.csv"), sep = ";", names = ["ayto", "contrato", "codigo"], encoding="latin1")
    contrats.contrato = contrats.contrato.apply(lambda x : str(x).replace("\n", ""))
    contrats = contrats[contrats.ayto == ayto]
    contrats = contrats[contrats.contrato == contractcode]
    display(contrats)
    if contrats.shape[0]!=1:
        print(contrats)
        print(ayto,"----",contractcode)
        return False
    return contrats.iloc[0]['codigo']
```
